# create and plot a random walk


```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
np.random.seed(42)
steps=np.random.standard_normal(1000)
steps[0]=0
random_walk=np.cumsum(steps)
```

```python
fig, ax = plt.subplots()

ax.plot(random_walk)
ax.set_xlabel('Timesteps')
ax.set_ylabel('Value')
```

```python
 #Create the differenced series (first differences)
diff_series = np.diff(random_walk)
```

```python
plt.hist(diff_series, bins=30, edgecolor='k', alpha=0.7, density=True)
plt.axvline(diff_series.mean(), color='red', linestyle='dashed', linewidth=2, label=f"Mean = {diff_series.mean():.4f}")
plt.axvline(diff_series.std(), color='blue', linestyle='dashed', linewidth=2, label=f"Std Dev = {diff_series.std():.4f}")
plt.axvline(-diff_series.std(), color='blue', linestyle='dashed', linewidth=2)
plt.legend()
plt.title("Histogram with Larger Sample Size")
plt.show()
```
